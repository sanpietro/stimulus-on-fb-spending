## Model a 12-month forecast for Total, Food, Drink & Dessert (Using FB Prophet)

```python
from fbprophet import Prophet
```

```python
d=[dft,dff,dfdr,dfdes]

for i in d:
    i.rename(columns = {'date':'ds','amount':'y'}, inplace = True)
```

```python
l1=[]
for i in d:
          
    m = Prophet()
    m.fit(i)
    future = m.make_future_dataframe(periods=12, freq='M')
    forecast = m.predict(future)
    l1.append(forecast)
```

```python
dftfcst = l1[0]
dfffcst = l1[1]
dfdrfcst = l1[2]
dfdesfcst = l1[3]
```

```python
fig = m.plot(dftfcst)
```

### Fairly good R^2, but does not reflect COVID spending changes

```python
metric_df = dftfcst.set_index('ds')[['yhat']].join(dft.set_index('ds').y).reset_index()

metric_df.dropna(inplace=True)

r2_score(metric_df.y, metric_df.yhat)
```

## Next, we take each category, model the forecast to get 2020 values, 
## adjust the values to account for stimulus spending beginning in April,
## and re-run the models to forecast 2021 months

## Food

```python
dfffcst = dfffcst[['ds','yhat']]
```

```python
dfffcst.head()
```

```python
from datetime import timedelta, date
start=dfffcst[(dfffcst['ds']>"2020-04-01") & (dfffcst['ds']<"2020-12-01")]
adj=dfffcst[(dfffcst['ds']<"2020-04-01") & (dfffcst['ds']>="2017-01-01")]
```

```python
# We model a 10% increase in food sales beginning in April

for d in start.ds:
    start['adjfcst']=(start['yhat']*1.1)

for x in adj.ds:
    adj['adjfcst']=(adj['yhat'])
```

```python
foodfcst=pd.concat([start, adj])

foodfcst.sort_values(by='ds')
```

```python
foodfcst.rename(columns = {'adjfcst':'y'},inplace=True)
foodfcst.drop(['yhat'],axis=1, inplace=True)
```

```python
f = Prophet()
f.fit(foodfcst)
future = f.make_future_dataframe(periods=12, freq='M')
forecast = f.predict(future)
fig = f.plot_components(forecast)
fig = f.plot(forecast)
```

```python
metric_df = forecast.set_index('ds')[['yhat']].join(foodfcst.set_index('ds').y).reset_index()

metric_df.dropna(inplace=True)

r2_score(metric_df.y, metric_df.yhat)
```

## Drink

```python
dfdrfcst= dfdrfcst[['ds','yhat']]
```

```python
start=dfdrfcst[(dfdrfcst['ds']>"2020-04-01") & (dfdrfcst['ds']<"2020-12-01")]
adj=dfdrfcst[(dfdrfcst['ds']<"2020-04-01") & (dfdrfcst['ds']>="2017-01-01")]
```

```python
# We model a 5% increase in drink sales beginning in April

for d in start.ds:
    start['adjfcst']=(start['yhat']*1.05)
for x in adj.ds:
    adj['adjfcst']=(adj['yhat'])
```

```python
drinkfcst=pd.concat([start, adj])

drinkfcst.sort_values(by='ds').head()
```

```python
drinkfcst.rename(columns = {'adjfcst':'y'},inplace=True)
drinkfcst.drop(['yhat'],axis=1, inplace=True)
drinkfcst.head()
```

```python
dr = Prophet()
dr.fit(drinkfcst)
future = dr.make_future_dataframe(periods=12, freq='M')
forecast = dr.predict(future)
fig = dr.plot_components(forecast)
fig = dr.plot(forecast)
```

```python
metric_df = forecast.set_index('ds')[['yhat']].join(drinkfcst.set_index('ds').y).reset_index()

metric_df.dropna(inplace=True)

r2_score(metric_df.y, metric_df.yhat)
```

## Dessert

```python
dfdesfcst= dfdesfcst[['ds','yhat']]
```

```python
start=dfdesfcst[(dfdrfcst['ds']>"2020-04-01") & (dfdesfcst['ds']<"2020-12-01")]
adj=dfdesfcst[(dfdesfcst['ds']<"2020-04-01") & (dfdesfcst['ds']>="2017-01-01")]
```

```python
start.head()
```

```python
# We model a 5% decline in dessert sales beginning in April

for d in start.ds:
    start['adjfcst']=(start['yhat']-(start['yhat']*.05))
for x in adj.ds:
    adj['adjfcst']=(adj['yhat'])
```

```python
dessertfcst=pd.concat([start, adj])
```

```python
dessertfcst.rename(columns = {'adjfcst':'y'},inplace=True)
dessertfcst.drop(['yhat'],axis=1, inplace=True)
dessertfcst.tail()
```

```python
des = Prophet()
des.fit(dessertfcst)
future = des.make_future_dataframe(periods=12, freq='M')
forecast = des.predict(future)
fig = des.plot_components(forecast)
fig = des.plot(forecast)
```

```python
forecast=forecast[forecast['yhat']>0]
```

```python
metric_df = forecast.set_index('ds')[['yhat']].join(dessertfcst.set_index('ds').y).reset_index()
```

```python
metric_df.dropna(inplace=True)
```

```python
r2_score(metric_df.y, metric_df.yhat)
```
